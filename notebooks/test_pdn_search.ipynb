Can't search VNAR sequences, won't be recognised as acceptable input

#### Search PLAbDab-nano by sequence

##### VHH search (with KA-search)

You can only search with VHH sequences or it will break

```python
vhh_query = 'QVQLQESGGGLVQAGGSLRLSCAASGTISYTPDMGWYRQAPGKEREFVAGITVGTSTYYADSVKGRFTISRDNAKNTVYLQMNSLKPEDTAVYYCAASRQWGPGFYYWGQGTQVTVSS'
```

```python
best_ident = pdn.vhh_seq_search({"H":vhh_query}, keep_best_n = 10, sort_by = 'average', regions=['whole'], length_matched=['Any'])
```

```python
best_ident
```

##### VNAR search (with BLAST)

You can only search with a VNAR sequence or nothing will be returned, as database to search against is built from only the existing VNAR sequences

```python
vnar_query = 'ARVDQTPRSVTKETGESLTINCVLRDASYALGSTCWYRKKSGSTNEESISKGGRYVETVNSGSKSFSLRINDLTVEDGGTYRCGVCVLVDYCDVLGCMRRWHCRDCECGLRGPPSPSPWRRLQGRRRQVGIQRYQSBNCBKL'
```

```python
pdn_vnar['sequence'][1]
```

Not length matched, whole sequence

```python
best_ident = pdn.vnar_seq_search(vnar_query, data_directory='path/to/db/', keep_best_n = 10, regions='whole', length_matched=False, url=True)
best_ident
```

Length matched, whole sequence

```python
best_ident = pdn.vnar_seq_search(vnar_query, data_directory='path/to/db/', keep_best_n = 10, regions='whole', length_matched=True, url=True)
best_ident
```

Length matched, CDR3 only 

```python
best_ident = pdn.vnar_seq_search(vnar_query, data_directory='path/to/db/', keep_best_n = 10, regions='cdr3', length_matched=True, url=True)
best_ident
```
